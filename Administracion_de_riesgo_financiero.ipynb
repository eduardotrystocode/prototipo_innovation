#**Black & Scholes**

El valor cuota se ve representado por el valor UF del día para una opción de compra futura en un plazo de vencimiento de dos años. Se espera un retorno establecido del ajuste de UF más un 4.6% de dividendos anualmente para el valor de la cuota (que sería nuestro derivado) al término de los dos años. El costo de la prima es el porcentaje de administración del fondo el cual corresponde a un 1.19%. Ahora bien, modelando la opción de compra desde el inicio de emisiones de las cuotas tenemos que nuestro valor S = $26,310.1 CLP, 

el cual corresponde al valor de nuestra cuota “actual” al término del primer trimestre del 2022. Ahora bien nuestro valor de ejercicio viene dado por el supuesto de la proyección del crecimiento del fondo en relación a los valores cuota, los cuales para el término del primer trimestre de 2024 corresponden a $28,786.301. El único valor que nos falta para poder valorar nuestra opción de compra de nuestro activo subyacente (el cual vendría siendo nuestro contrato de compra de las cuotas de participación del fondo inmobiliario LV - Patio I) corresponde a la volatilidad del mismo, es decir la desviación de la rentabilidad del fondo. Tomando como base la rentabilidad de los 6 años previos a 2022, el resultado de las desviación de nuestro activo nos arroja un sigma de 3,88%. Esta volatilidad es baja debido a que el fondo elegido es respaldado por la UF, lo cual actúa como un escudo contra la inflación y los rendimientos decrecientes.

Ahora bien, con todos los parámetros dados podemos calcular mediante el modelo de Black & Scholes la valoración de la opción de compra correspondiente al fondo inmobiliario escogido.


```python
import numpy as np
import scipy.stats as si
import sympy as sy
#import sympy.statistics as systats
```

```python
def valoracion_blackscholes(S, K, T, r, sigma):
    # Modelo BS esta puesto en años
    #S: Precio del activo
    #K: Precio de ejercicio
    #T: Tiempo para vencimiento
    #r: Tasa de interes
    #sigma: volatibilidad del activo
    
    d1 = (np.log(S / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = (np.log(S / K) + (r - 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    
    call = (S * si.norm.cdf(d1, 0.0, 1.0) - K * np.exp(-r * T) * si.norm.cdf(d2, 0.0, 1.0))
    
    return call
```

```python
valoracion_blackscholes(26310.1, 28786.301, 2, 0.0119, 0.0388)
```

Podemos ver que la opción de compra tiene un valor DE $82.7588 CLP el cual es excesivamente bajo, pues la variación del valor cuota para el año 2024 es mínima según las proyecciones del precio de ejercicio. Esto se puede explicar debido a los rendimientos poco volátiles del fondo comparado con el crecimiento de este.


#**Modelo adicional**

## Teoria

Las hipótesis básicas o de partida del método binomial, que se trata de un modelo discreto, son las siguientes:

1. Se suponen mercados de capitales competitivos y perfectos, en los que no hay impuestos, costes de transacción o limitaciones a la operativa en descubierto.
2. No existen restricciones sobre compras o ventas a corto plazo de títulos y opciones, por lo que el volumen de transacciones no afectará al precio de mercado de los títulos.
3. Existe una tasa de interés a corto plazo sin riesgo, rf, conocida, positiva y constante a lo largo del período considerado, lo que implica la posibilidad de prestar o pedir prestado dinero a ese tipo, o lo que es lo mismo, se supone la existencia de un bono libre de riesgo cuyo rendimiento es precisamente rf.
4. El horizonte de planificación está dividido en períodos de tiempo iguales (tiempo discreto).
5. La acción o activo subyacente no paga dividendos ni otro tipo de distribución de beneficios, reservas o capital.
6. El precio de la acción sigue un proceso binomial multiplicativo a lo largo de períodos discretos de tiempo.
7. La tasa de retorno de la acción solo puede tomar en cada momento dado del tiempo dos posibles valores: u-1 con probabilidad q y d-1 con probabilidad 1-q, (u>d).

Fórmula Binomial de valoración del precio de una opción de compra CALL:

C = S · Φ(j > a;a,p) - E · r-n · Φ(j > a;n,p´)

donde:

c = prima de la opción de compra.

S = precio actual de la acción.

E = precio de ejercicio de la opción.

r = uno más la tasa de interés sin riesgo.

n = nº de períodos que faltan para la expiración de la opción.

j = nº de movimientos ascendentes del precio de la acción (n-j movimientos descendentes).

Φ = complementaria de la función de distribución binomial (coeficiente corrector por la existencia de riesgo), que tiene la siguiente expresión:



A continuacion se resuelve el problema presentado en el insiso anterior, esta vez utilizando un modelo binomial:


```python
#codigo obtenido de https://github.com/financieras/derivados/blob/main/ModeloBinomial.ipynb

import numpy as np
from scipy.stats import binom
def prima(tipo,spot,strik,anyos,volatilidad,n,tInt):
  r=np.log(1+tInt)  # r es el tanto instantáneo
  deltaT=anyos/n
  u=np.exp(volatilidad*np.sqrt(deltaT))
  d=1/u
  p=(np.exp(r*deltaT)-d)/(u-d)
  Sn=[0]*(n+1) #Sn es la lista de precios del activo subyacene a fecha de vto.
  payoff=[0]*(n+1) #inicilizamos a cero la lista del payoff
  prima=0
  for i in range(n+1):
    Sn[i]=spot*u**(n-i)*d**i
    if tipo.lower()=='call':
      payoff[i]=max(0,Sn[i]-strik)
    if tipo.lower()=='put':
      payoff[i]=max(0,strik-Sn[i])
    prima+=payoff[i]*binom.pmf(n-i, n, p)
  prima*=np.exp(-r*anyos) #descontamos a valor actual, hasta t=0
  return prima

m=500 #Calcularemos el arbol desde 1 hasta m pasos
evolucion=[0]*m
for j in range(m): #se pone j+1 pq las listas empiezan en cero
  evolucion[j]=prima('call',26310.1, 28786.301,2,0.0388,j+1,0.0119)
  print('prima(',j+1,')=',evolucion[j])
```

Como se puede observar, luego de 500 iteraciones, el valor tiende a 82.2132 CLP. Un valor similar al obtenido por Black & Scholes. Tal como se dijo anteriormente, el bajo valor puede atribuirse al bajo nivel de volatilidad que precenta la opcion.

#**Análisis**

Ambos modelos entregan un valor de 82 CLP, diferenciandose solamente en los valores decimales.

82 CLP representa solo un 0.28% del valor de *strike*. Como se declaro en la seccion de Derivados, las opciones de compra se usan para asegurarse contra los distintos riesgos. Los derivados son una herramienta para manejar el riesgo financiero. Uno de los riesgos mas importantes es la volatilidad del valor del activo. Por esto mismo, al tener una volatilidad baja, el valor de la opcion tambien resulta bajo.
